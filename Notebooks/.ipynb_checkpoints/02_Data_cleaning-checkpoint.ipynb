# Data Cleaning – Telecom Customer Churn

## Purpose
The objective of this notebook is to clean and preprocess the raw telecom customer dataset by handling incorrect data types, missing values, and irrelevant features to prepare the data for exploratory data analysis and machine learning.

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r'C:\Users\AKSHAT\Telecom_Churn_Project/Data/Telco-Customer-Churn.csv')
```

```python
df.head()
```

```python
df.columns
```

```python
df.dtypes
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'],errors='coerce')
```

```python
### missing values
df.isnull().sum()
```

```python
df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

```python
df.drop(columns=['customerID'],inplace=True)
```

```python
### saveing cleaned dataset
df.to_csv(r'C:\Users\AKSHAT\Telecom_Churn_Project/Data/Cleaned_dataset.csv')
```
