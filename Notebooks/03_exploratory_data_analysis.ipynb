# Exploratory Data Analysis (EDA) – Telecom Customer Churn

## Purpose
The objective of this notebook is to explore customer churn patterns, identify key factors influencing churn, and generate business insights that will guide feature engineering, machine learning modeling, and dashboard creation.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style='whitegrid')
```

```python
df = pd.read_csv(r'C:\Users\AKSHAT\Telecom_Churn_Project\Data\Cleaned_dataset.csv')
```

```python
df.head()
```

```python
### overall churn rate
churn_rate = df['Churn'].value_counts(normalize=True)*100
churn_rate
```

Approximately 26.536897% of customers have churned, indicating a significant retention challenge for the telecom company.

```python
### churn by contract type 

plt.figure(figsize=(8,5))
sns.countplot(x='Contract',hue='Churn',data=df)
plt.title('Churn by Contrcat Type')
```

 Customers on month-to-month contracts show significantly higher churn compared to long-term contracts.

```python
### Churn VS Tenure

plt.figure(figsize=(8,5))
sns.histplot(data=df,x='tenure',hue='Churn',bins=30,kde=True)
plt.title('Churn distribution by tenures')
plt.show()
```

Customers with shorter tenure are more likely to churn, suggesting onboarding and early engagement issues.


```python
### monthly charges vs Churn
plt.figure(figsize=(8,6))
sns.boxplot(x='Churn',y='MonthlyCharges',data=df)
plt.title('Monthly charges vs churn')
plt.show()
```

Customers who churn tend to have higher monthly charges, indicating price sensitivity.

```python
### Churn b y internet service
plt.figure(figsize=(8,6))
sns.countplot(x='InternetService',hue='Churn',data=df)
plt.title('Churn By internet service Type')
plt.show()
```

Fiber optic customers show higher churn compared to other internet services.

```python
### churn by payment method
plt.figure(figsize=(10,5))
sns.countplot(y='PaymentMethod',hue='Churn',data=df)
plt.title('Churn by Payment Method')
plt.show()
```

Customers using electronic check payment methods exhibit higher churn rates.

### Key Insights from EDA
Month to month contracts have the highest churn
customers with low tenure are at a higher risk
customers with higher monthly charges tend to churn
fiber optic users churn more
payment mehtod influences churn behaviour
