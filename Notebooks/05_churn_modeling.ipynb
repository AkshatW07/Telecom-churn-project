# Churn Prediction Modeling – Telecom Customer Churn

## Purpose
The objective of this notebook is to build a baseline machine learning model to predict customer churn, evaluate its performance using appropriate metrics, and interpret results to support business decision-making.


```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    confusion_matrix,
    classification_report,
    roc_auc_score
)
```

```python
df = pd.read_csv(r'C:\Users\AKSHAT\Telecom_Churn_Project DA\Data\churn_features.csv')
df.head()
```

```python
x = df.drop('Churn',axis=1)
y = df['Churn']
```

```python
###train-test split
x_train, x_test, y_train, y_test = train_test_split(
    x,
    y,
    test_size=0.2,
    random_state=42,
    stratify=y
)
```

```python
###Train Logistic Regression Model
log_reg = LogisticRegression(max_iter=1000)
log_reg.fit(x_train, y_train)
```

```python
###predictions
y_pred = log_reg.predict(x_test)
y_pred_prob = log_reg.predict_proba(x_test)[:, 1]
```

```python
###model evalution
accuracy_score(y_test, y_pred)
```

```python
###confusion matrix
confusion_matrix(y_test, y_pred)
```

```python
###classification report
print(classification_report(y_test, y_pred))
```

```python
###roc curve
roc_auc_score(y_test, y_pred_prob)
```

```python
###Interpret Model Coefficients
coefficients = pd.DataFrame({
    'Feature': x.columns,
    'Coefficient': log_reg.coef_[0]
}).sort_values(by='Coefficient', ascending=False)

coefficients.head(10)
```

## Model Insights
- Contract type is one of the strongest churn predictors
- Customers with higher monthly charges have higher churn risk
- Longer tenure significantly reduces churn probability
- Certain services and payment methods influence churn behavior
